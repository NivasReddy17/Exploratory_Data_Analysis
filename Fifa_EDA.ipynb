# Comparision of nations and players

```python
counts_Nationality = fifa["Nationality"].value_counts()
counts_Nationality = counts_Nationality.reset_index()
counts_Nationality.columns= ["Nations","Counts"]
counts_Nationality
```

```python
plt.figure(figsize = (10, 35))
sns.barplot(x = "Counts", y = "Nations", data = counts_Nationality)
plt.title("Nation Wise Players Counts");
plt.show()
```

# Comparision between number of players and positions

```python
plt.figure(figsize=(15,5))
sns.countplot('Position', data=fifa)
plt.xlabel("Positions")
plt.ylabel("Number of players")
plt.title("Representation of players in different positions")
plt.show()
```

# Players Overall

```python
plt.figure(figsize=(15,5))
sns.countplot('Overall', data=fifa)
plt.xlabel("Overall Rating")
plt.ylabel("Number of players")
plt.title("Overall rating distribution")
plt.show()
```

# Potential Distribution

# Foot Preference

```python
foot_preference = fifa['Preferred Foot'].value_counts().reset_index()
foot_preference.columns = ['Foot', "Total Players"]
foot_preference = foot_preference.set_index('Foot')
foot_preference
```

```python
plt.figure(figsize = (5,5))
sns.countplot('Preferred Foot', data= fifa)
plt.xlabel("Foot preference")
plt.ylabel("Number of players")
plt.title("Preferred Foot")
plt.show()
```

# Clubs

```python
highest_overall_club = fifa.groupby('Club').Overall.mean().reset_index().sort_values(by='Overall', ascending=False)
highest_overall_club
```

```python
top_clubs = highest_overall_club.head(15)
plt.figure(figsize = (10,5))
sns.barplot(x=top_clubs.Overall, y=top_clubs['Club'])
plt.title("Top 15 Best Overall Clubs");
```

# Clubs with highest Overall

```python
plt.figure(figsize=(12,8)) 
ax = fifa.groupby(['Club'])['Overall'].max().sort_values(ascending = False).head(10).plot.bar()
ax.set_xlabel('Clubs')
ax.set_ylabel('Potential')
ax.set_title("Clubs with highest Overall")
plt.show()
```

# Clubs with highest Potential

```python
plt.figure(figsize=(12,8))
ax = fifa.groupby(['Club'])['Potential'].max().sort_values(ascending = False).head(15).plot.bar()
ax.set_xlabel('Clubs')
ax.set_ylabel('Potential')
ax.set_title("Clubs with highest Potential score")
plt.show()
```
